# Η Λογική Πολυπλοκότητα της Συνεργασίας

Ένα από τα βασικά χαρακτηριστικά του ανθρώπινου είδους είναι η ικανότητά των μελών του είδους να συνεργάζονται μεταξύ τους. Αν και σε άλλα είδη εμφανίζονται φαινόμενα συνεργασίας, αυτά εντοπίζονται κυρίως σε επίπεδο οικογένειας ή κοντινής ομάδας. Οι άνθρωποι όμως μπορούμε να συνεργαζόμαστε σε ομάδες χιλιάδων ατόμων, αγνώστων μεταξύ τους. 

Ένα ερώτημα τότε είναι, πώς προκύπτουν σχέσεις συνεργασίας μεταξύ ατόμων; Για να μελετήσουμε τις συνθήκες κάτω από τις οποίες επιλέγουμε να συνεργαστούμε, ή αντιστρόφως, επιλέγουμε να μη συνεργαστούμε, εργαζόμαστε με συγκεκριμένα μοντέλα. Στα μοντέλα αυτά θεωρούμε ότι αποτυπώνουμε την αλληλεπίδραση μεταξύ δύο ατόμων κάθε φορά. Στο ζευγάρι αυτό κάθε άτομο έχει ένα συγκεκριμένο ρόλο: υπάρχει ένας *δωρητής* (donor) και ένας *παραλήπτης* (recipient). Η ιδέα είναι ότι ο δωρητής έχει τη δυνατότητα να δώσει (ή να μη δώσει) κάτι στον παραλήπτη. Αυτό το κάτι μπορεί να είναι ένα υλικό αγαθό, ή μπορεί να είναι αρωγή σε κάποια προσπάθεια. Σε κάθε περίπτωση, ο δωρητής έχει ένα κόστος όταν αποφασίζει να συνδράμει τον παραλήπτη, συνεπώς, όταν το κάνει, κάνει μία συνειδητή επιλογή συνεργασίας.

Ο δωρητής στο μοντέλο μας έχει μια *φήμη* (reputation), η οποία εξαρτάται από το πώς έχει φερθεί στο παρελθόν, σε προηγούμενες περιπτώσεις. Επίσης, φήμη έχει και ο παραλήπτης. Ο παραλήπτης μάλιστα έχει δύο φήμες: αυτή που έχει αυτή τη στιγμή (βάσει φυσικά της συμπεριφοράς του στο παρελθόν), και αυτή που είχε προηγουμένως στο παρελθόν. Άρα ένας παραλήπτης μπορεί τώρα να έχει καλή φήμη, αλλά να ξέρουμε ότι παλαιότερα δεν είχε.

Όταν προκύπτει περίπτωση συνεργασίας μεταξύ ενός πιθανού δωρητή και παραλήπτη, ο δωρητής μπορεί να επιλέξει να βοηθήσει ή όχι τον παραλήπτη, λαμβάνοντας υπόψη την τρέχουσα και την προηγούμενη φήμη του. Ανάλογα με το τι θα κάνει, θα επηρεαστεί η δικιά του φήμη. Για παράδειγμα, είναι λογικό ότι ένας δωρητής που αρνείται να βοηθήσει κάποιον που έχει καλή φήμη, θα αποκτήσει κακή φήμη, καθώς θα ληφθεί ως εγωιστής.

## Νόρμες Αξιολόγησης

Έχοντας το δωρητή και τον παραλήπτη, θέλουμε να αποτυπώσουμε τους τρόπους με τους οποίους η φήμη του δωρητή θα επηρεαστεί από την επιλογή που θα κάνει σε μία συναλλαγή με τον παραλήπτη. Έχουμε τέσσερεις παράγοντες που θα πρέπει να λάβουμε υπόψη μας:

* Την επιλογή του δωρητή, να βοηθήσει ή να μην βοηθήσει.
* Την τρέχουσα φήμη του δωρητή, καλή ή κακή.
* Την τρέχουσα φήμη του παραλήπτη, καλή ή κακή.
* Την προηγούμενη φήμη του παραλήπτη, καλή ή κακή.

Μπορούμε να αντιστοιχίσουμε μία δυαδική μεταβλητή σε κάθε έναν από τους παραπάνω παράγοντες, ως εξής:

* $A = 1$ αν ο δωρητής επιλέξει να βοηθήσει, $A = 0$ αν αγνοήσει την επίκληση για βοήθεια.
* $B = 1$ αν ο δωρητής έχει καλή φήμη, $B = 0$ αν δεν έχει.
* $C = 1$ αν ο παραλήπτης έχει καλή φήμη, $C = 0$ αν δεν έχει.
* $D = 1$ αν ο παραλήπτης είχε καλή φήμη, $D = 0$ αν δεν είχε.

Έτσι, έχουμε $2^4 = 16$ διαφορετικούς συνδυασμούς των $A$, $B$, $C$, $D$ που περιγράφουν μια συναλλαγή μεταξύ δωρητή και παραλήπτη, όπως φαίνεται στο ακόλουθο σχήμα:

<img src="empty_map.png" width="400px">

Σε κάθε ένα συνδυασμό τώρα, μπορούμε να δώσουμε την τιμή 1, αν ως αποτέλεσμα της συναλλαγής η φήμη του δωρητή θα είναι καλή, και την τιμή 0 αν ως αποτέλεσμα της συναλλαγής η φήμη του δωρητή θα είναι κακή. Με αυτόν τον τρόπο γεμίζει το σχήμα με 0 και 1. Δείτε το παρακάτω παράδειγμα:

<img src="image_score_no_implicant.png" width="400px">

Αν προσέξετε αυτόν τον πίνακα, θα δείτε ότι σημαίνει ότι η φήμη του δωρητή μετά τη συναλλαγή θα είναι ακριβώς η επιλογή του: θα έχει καλή φήμη αν βοηθήσει και κακή φήμη αν δεν βοηθήσει, ανεξαρτήτως της υπάρχουσας φήμης του ή της φήμης που έχει και είχε ο παραλήπτης.

Κάθε ένας τρόπος να γεμίσει το τετράγωνο ορίζει έτσι μια κοινωνική *νόρμα* (norm) με την οποία αξιολογείται το ποιόν του δωρητή. Υπάρχουν πολλές τέτοιες δυνατές νόρμες: $2^{16}$. Σκοπός μας λοιπόν είναι ακριβώς να μπορούμε να μελετήσουμε τις νόρμες αυτές.

## Απλοποίηση Νόρμας

Όταν περιγράφεται μια νόρμα με τη μορφή του πίνακα, δεν είναι εύκολο να καταλάβουμε τι ακριβώς σημαίνει. Για το λόγο αυτό, θέλουμε ναμπορούμε να απλοποιήσουμε μια νόρμα στην πιο απλή δυαδική έκφραση που να την περιγράφει. Αν δούμε πάλι το προηγούμενο παράδειγμα, μπορούμε να παρατήσουμε ότι εφ' όσον το μόνο που έχει σημασία είναι το τι θα πράξει ο δωρητής, η νόρμα περιγράφεται από την απλή σχέση:

$$ F = A $$

Αυτό το δείχνουμε και στην παρακάτω εικόνα, όπου έχουμε σκιάσει το μέρος της νόρμας που έχει τιμή 1, και παρατηρούμε ότι περιγράφεται ακριβώς από την τιμή $A = 1$, ανεξαρτήτως του τι τιμές έχουν οι άλλες μεταβλητές:

<img src="image_score.png" width="400px">

Η νόρμα αυτή ονομάζεται *είδωλο* (image), γιατί το αποτέλεσμά της είναι το είδωλο της πράξης του δωρητή.

Δείτε τώρα την εξής νόρμα:

<img src="simple_standing.png" width="400px">

Εδώ η νόρμα δίνει την τιμή 1 αν και μόνο αν $A = 1$ ή η μεταβλητή $C$ είναι ψευδής, $\overline{C} = 1$, άρα περιγράφεται από τη σχέση: $ F = A \lor \overline{C} $. Η σχέση σημαίνει ότι ο δωρητής θα λάβει καλή φήμη αν βοηθήσει ή αν ο παραλήπτης έχει κακή φήμη, και ονομάζεται *απλή υπόληψη* (simple standing). Η ονομασία εξηγείται αν σκεφτούμε ότι, αντιστρόφως, η υπόληψη (standing) του δωρητή μειώνεται απλώς αν και μόνο αν δεν βοηθήσει κάποιον με καλή υπόληψη. Το σύμβολο $\lor$ είναι το λογικό Ή (διάζευξη) και η παύλα πάνω από μία μεταβλητή σημαίνει την άρνησή της.

Μπορεί όμως να είμαστε πιο αυστηροί στην κρίση μας, όπως στην παρακάτω νόρμα:

<img src="stern_judging.png" width="400px">

Εδώ ο δωρητής κρίνεται θετικά αν βοηθήσει κάποιον με καλή φήμη ή αρνηθεί να βοηθήσει κάποιον με κακή φήμη. Συνεπώς η νόρμα ονομάζεται *αυστηρή κρίση* (stern judging). Στις παραστάσεις που γράφουμε, όταν δύο μεταβλητές βρίσκονται δίπλα η μία στην άλλη, όπως στην $AC$, σημαίνει ότι συνδέονται με το λογικό ΚΑΙ (σύζευξη).

Σε περίπτωση που για να κρίνουμε τον δωρητή θετικά θέλουμε, σε περίπτωση που αρνηθεί να βοηθήσει, να το κάνει αυτό έχοντας καλή φήμη, φτάνουμε στη νόρμα που ονομάζεται *κρίση* (judging):

<img src="judging.png" width="400px">


Μία νόρμα χαρακτηρίζεται από την πολυπλοκότητά της. Όπως είδαμε, κάποιες περιγράφονται από πιο απλές σχέσεις από ό,τι άλλες. Αυτό μετριέται με τη *λογική πολυπλοκότητα* (boolean complexity), η οποία είναι ο αριθμός των εμφανίσεων μεταβλητών στη σχέση. Έτσι, η σχέση $F = AC \lor \overline{A}B\overline{C}$ έχει πολυπλοκότητα 5.

### Εύρεση Λογικών Όρων από Ομάδες

Κάθε ομάδα αντιστοιχεί σε έναν λογικό όρο, τον οποίο μπορούμε να βρούμε εντοπίζοντας τις τιμές των μεταβλητών που μένουν αμετάβλητες σε όλα τα κελιά της ομάδας. Οι τιμές αυτές μας δίνουν τον όρο που ψάχνουμε. Ας δούμε για παράδειγμα τη νόρμα της κρίσης:

Η ομάδα του τετραγώνου αποτελείται από τα κελιά:

<img src="rectangle_term.png" width="125px">

Όπως βλέπουμε, η πρώτη και η τρίτη στήλη έχουν παντού την ίδια τιμή, 1. Άρα η ομάδα αντιστοιχεί στη λογική παράσταση που απαιτεί $A = 1$  $C = 1$, δηλαδή $AC$.

Η ομάδα της στήλης αποτελείται από τα κελιά:

<img src="hpair_term.png" width="125px">

Όπως βλέπουμε, η πρώτη, η δεύτερη, και η τρίτη στήλη έχουν παντού την ίδια τιμή, 0, 1, και 0 αντίστοιχα. Άρα η ομάδα αντιστοιχεί στη λογική παράσταση που απαιτεί $A = 0$  $B = 1$, $C = 0$ δηλαδή $\overline{A}B\overline{C}$.

Η λογική παράσταση που προκύπτει από κάθε ομάδα αποτελείται από μία ή παραπάνω μεταβλητές συνδεμένες μεταξύ τους με ένα λογικό και (σύζευξη). Τέτοιες λογικές παραστάσεις ονομάζονται *ελαχιστόροι* (minterms).

Η διαδικασία εύρεσης των ελαχιστόρων εξηγεί γιατί οι συντεταγμένες του τετραγώνου ακολουθούν τη σειρά 00, 01, 11, 10 στο δυαδικό, ή 0, 1, 3, 2 αντί για 0, 1, 2, 3 που μπορεί να περίμενε κανείς. Η σειρά αυτή είναι ένα παράδειγμα [κώδικα Gray](https://en.wikipedia.org/wiki/Gray_code): μια σειρά δυαδικών αριθμών που ο ένας διαφέρει από τον άλλον κατά ένα bit. Χάρη σε αυτήν την ιδιότητα της ακολουθίας οι ομάδες των ελαχιστόρων έχουν κοινές τιμές σε θέσεις των bits.

### Ένωση Λογικών Εκφράσεων

Το τελικό αποτέλεσμα προκύπτει από τη λογική διάζευξη (Ή) των ελαχιστόρων που βρήκαμε, δηλαδή $AC \lor \overline{A}B\overline{C}$.

Γενικότερα ισχύει το εξής:

* Μια ομάδα των 16 κελιών περιγράφεται από την τιμή αλήθειας (1) ή ψεύδους (0).
* Μια ομάδα των 8 κελιών περιγράφεται από μία λογική μεταβλητή.
* Μια ομάδα των 4 κελιών περιγράφεται από δύο λογικές μεταβλητές.
* Μια ομάδα των 2 κελιών περιγράφεται από τρεις λογικές μεταβλητές.
* Ένα κελί μόνο του περιγράφεται από τέσσερεις λογικές μεταβλητές (οι συντεταγμένες του ενός κελιού).

### Λεπτομέρειες Εξόδου

Επειδή δεν μπορούμε να γράψουμε εύκολα παύλες πάνω από τους χαρακτήρες στο τερματικό μας, θα χρησιμοποιούμε το χαρακτήρα `~`, δηλαδή αντί για $\overline{A}$ θα γράφουμε `~A`. Ο χαρακτήρας $\lor$ μπορεί να τυπωθεί με τον χαρακτήρα 2228 του Unicode, για παράδειγμα με `print('\u2228')`.

Οι ελαχιστόροι θα πρέπει να εμφανίζονται με την εξής σειρά:

1. Αύξουσα σειρά μήκους, όπου το μήκος αντιστοιχεί στο πλήθος των μεταβλητών του κάθε ελαχιστόρου.

2. Σε περίπτωση ίδιου μήκους, οι ελαχιστόροι με το ίδιο μήκος θα διατάσσονται αλφαβητικά, όπου θεωρούμε ότι το `~X` έπεται του `X`.

## Περισσότερες Πληροφορίες

Το πρόβλημα που θέσαμε ξεκίνησε από τη μελέτη κανόνων συνεργασίας, αλλά ο τρόπος με τον οποίο απλοποιούμε τις κοινωνικές νόρμες δεν είναι άλλος από τη χρήση [χαρτών Karnaugh](https://en.wikipedia.org/wiki/Karnaugh_map). Οι χάρτες Karnaugh χρησιμοποιούνται στην απλοποίηση λογικών παραστάσεων, όπως αυτές που απαντάμε στη σχεδίαση ψηφιακών κυκλωμάτων.

Παρά το ότι με τους χάρτες Karnaugh μπορούμε να απλοποιήσουμε εύκολα λογικές παραστάσεις μέχρι κάποιο μικρό αριθμό μεταβλητών, στην πράξη συχνά χρησιμοποιείται o [αλγόριθμος Quine-McCluskey](https://en.wikipedia.org/wiki/Quine%E2%80%93McCluskey_algorithm). *Στην εργασία δεν επιτρέπεται η χρήση του αλγορίθμου αυτού*.

Ο αλγόριθμος Quine-McCluskey μπορεί να προβλήματα με περισσότερες λογικές μεταβλητές από ότι οι χάρτες Karnaugh (όπου σταματάμε στις τέσσερεις), αλλά και πάλι, δεν μπορούμε να απλοποιήσουμε παραστάσεις με μεγάλο αριθμό μεταβλητών. Το πρόβλημα ανήκει στην κατηγορία [NP-hard](https://en.wikipedia.org/wiki/NP-hardness).
